# WDBC part, done in class

```python
import pandas as pd

df = pd.read_csv('wdbc.data', header=None)
df.head()
```

```python
df.iloc[:,1].unique()
```

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error',
       'fractal dimension error', 'worst radius', 'worst texture',
       'worst perimeter', 'worst area', 'worst smoothness',
       'worst compactness', 'worst concavity', 'worst concave points',
       'worst symmetry', 'worst fractal dimension']
```

```python
df.columns = column_labels
df
```

```python
df['diagnosis'].replace('B', 'Benign', inplace=True)
df['diagnosis'].replace('M', 'Malignant', inplace=True)
df
```

```python
feature_df = df.iloc[:,2:]
label_df = df.iloc[:,1]
```

```python
# Use sklearn to standardise (specifically: normalise) the data so that 
# the data has a normal distribution -- mean = 0; stdev = 1
from sklearn.preprocessing import StandardScaler
import numpy as np

x = feature_df.values
x = StandardScaler().fit_transform(x)
# ... and check if it worked
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
normalised_breast
```

```python
# Choose the first 2 principal components
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)

principal_breast_df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2'])
```

```python
# One thing that's often useful is "how much of the variance is explained"
# in any principal component. We can find out that 44.28% is explained
# by PC1; 18.97 by PC2:
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
# Let's plot it!
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1',fontsize=20)
plt.ylabel('Principal Component 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_breast_df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

as we can see:

principal component 1 can explain more, so we can find the spot become red if principal_component_1 > some threshold

```python
pca = PCA().fit(x)
evr = pca.explained_variance_ratio_

print("percentage of variance explained: ")
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

# Gisette Part

## Read Data

```python
# import pandas and read data
import pandas as pd

df_g = pd.read_csv('gisette_train.data', header = None, delim_whitespace = True)

df_g
```

```python
# also read the test set
df_test = pd.read_csv('gisette_test.data', header = None, delim_whitespace = True)

df_test
```

```python
# and validation set
df_valid = pd.read_csv('gisette_valid.data', header = None, delim_whitespace = True)

df_valid
```

```python
# read train set labels
df_g_labels = pd.read_csv('gisette_train.labels', header = None)
df_g_labels
```

```python
# and validation set labels
df_valid_labels = pd.read_csv('gisette_valid.labels', header = None)
df_valid_labels
```

```python
df_g_labels[0].unique()
```

in labels, there are two kind of values, 1 and -1

## Normalise Data

```python
# normalise
from sklearn.preprocessing import StandardScaler
import numpy as np

x = df_g.values
x = StandardScaler().fit_transform(x)

x_test = df_test.values
x_test = StandardScaler().fit_transform(x_test)

x_valid = df_valid.values
x_valid = StandardScaler().fit_transform(x_valid)

# check if it is working?
print(np.mean(x), np.std(x))
print(np.mean(x_test), np.std(x_test))
print(np.mean(x_valid), np.std(x_valid))
```

```python
# change x back into a dataframe again, actually not necessary
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_gisette = pd.DataFrame(x,columns=feat_cols)
normalised_gisette.tail()
normalised_gisette
```

## PCA for all three sets

```python
# also choose the first 2 principal components
from sklearn.decomposition import PCA

pca_gisette = PCA(n_components=2)
principalComponents_gisette = pca_gisette.fit_transform(x)

# also apply this pca to test set and validation set
valid_pca = pca_gisette.transform(x_valid)
test_pca = pca_gisette.transform(x_test)

principal_gisette_df = pd.DataFrame(data = principalComponents_gisette
             , columns = ['principal component 1', 'principal component 2'])

valid_pca_df = pd.DataFrame(data = valid_pca
             , columns = ['principal component 1', 'principal component 2'])

test_pca_df = pd.DataFrame(data = test_pca
             , columns = ['principal component 1', 'principal component 2'])
```

```python
# print "how much of the variance is explained"

print('Explained variation per principal component: {}'.format(pca_gisette.explained_variance_ratio_))
```

```python
# plot function
import matplotlib.pyplot as plt

%matplotlib inline

def my_pca_plot(df, label, title):
    plt.figure()
    plt.figure(figsize=(10,10))
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=14)
    plt.xlabel('Principal Component 1',fontsize=20)
    plt.ylabel('Principal Component 2',fontsize=20)
    plt.title(title,fontsize=20)
    targets = [1, -1]
    colors = ['g', 'r']
    for target, color in zip(targets,colors):
        indicesToKeep = label == target
        plt.scatter(df.loc[indicesToKeep, 'principal component 1']
                   , df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

    plt.legend(targets,prop={'size': 15})
    
    
```

```python
# plot train set
my_pca_plot(principal_gisette_df, df_g_labels[0], "Principal Component Analysis of Gisette Training Dataset")
```

```python
# plot validation set
my_pca_plot(valid_pca_df, df_valid_labels[0], "Principal Component Analysis of Gisette Validation Dataset")
```

```python
# since we do not have labels for test set, we can only plot them without label
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1',fontsize=20)
plt.ylabel('Principal Component 2',fontsize=20)
plt.title("Principal Component Analysis of Gisette Test Dataset",fontsize=20)
plt.scatter(test_pca_df.loc[:, 'principal component 1']
            , test_pca_df.loc[:, 'principal component 2'], c = 'g', s = 50)
```

```python
pca = PCA().fit(x)
evr = pca.explained_variance_ratio_
```

```python
print("percentage of variance explained: ")
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

As we can see, principal component 1 and principal component 2 can not explain that much.

So, we should choose not only 2 features.

```python
# Graph it...
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```

As it shows in this graph,

I think choosing the first 2000 or 3000 features would be enough.
